# Machine Learning in Python- Scaling Data

### 1) Importing Libraries and Setting Path

```python
#import libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib
import matplotlib.pyplot as plt
import os
import sklearn
from sklearn.preprocessing import StandardScaler

#creating Paths
path= r'C:\Users\wjzak\Dropbox\Career Foundry\Data Specialization\Module 1\12-25- ClimateWins Project'

#using the path to pull file name and read the clean products data
df = pd.read_csv(os.path.join(path, 'Data Sets', 'Dataset-weather-prediction-dataset-processed.csv'), index_col = False)
```

### 2) Data Prep

```python
df.head()
```

```python
df.reset_index(inplace=True)
df.rename(columns={'index':'id'}, inplace=True)
df.head()
```

```python
df.isnull().sum()
```

```python
#eliminate those records that have NaN values

df.dropna(inplace=True)
```

### 3) Scaling

#create scaler object using StandardScaler from sklearn.preprocessing
#StandardScaler assumes data is normally distributed and scales with a distribution
#around 0 and standard deviation of 1. Scaling happens independently with each variable.

```python
# Scale only numeric columns, excluding id/DATE/MONTH
exclude = ['id', 'DATE', 'MONTH']
cols_to_scale = [c for c in df.columns if c not in exclude and pd.api.types.is_numeric_dtype(df[c])]
scaler = StandardScaler()
df_scaled = df.copy()
df_scaled[cols_to_scale] = scaler.fit_transform(df[cols_to_scale])
df_scaled.head()
```

### 5) Data Export to Prepared Folder

```python
#Save updated csv file to Prepared Data Folder
df_scaled.to_csv(os.path.join(path, 'Data Sets', 'Dataset-weather-prediction-scaled.csv'))
```
